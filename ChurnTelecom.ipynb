# Projeto de Previsão Customer Churn em Operadoras de Telecom

Customer Churn (ou Rotatividade de Clientes, em uma tradução livre) refere-se a uma decisão tomada pelo cliente sobre o término do relacionamento comercial. Refere-se também à perda de clientes. A fidelidade do cliente e a rotatividade de clientes sempre somam 100%. Se uma empresa tem uma taxa de fidelidade de 60%, então a taxa de perda de clientes é de 40%. 

De acordo com a regra de lucratividade do cliente 80/20, 20% dos clientes estão gerando 80% da receita. Portanto, é muito importante prever os usuários que provavelmente abandonarão o relacionamento comercial e os fatores que afetam as decisões do cliente.

Neste projeto, irei realizar a previsão do Customer Churn em uma Operadora de Telecom utilizando o algoritmo de Regressão Logística.

```python
# Carregando dados de teste
teste = pd.read_csv("data/projeto4_telecom_teste.csv")
teste.head()
```

```python
# Realizando mesmo procedimento com os dados de treino
teste = teste.drop(teste.columns[0],axis=1)
teste['state'] = label_encoder.fit_transform(teste['state'])
teste['area_code'] = label_encoder.fit_transform(teste['area_code'])
teste['international_plan'] = label_encoder.fit_transform(teste['international_plan'])
teste['voice_mail_plan'] = label_encoder.fit_transform(teste['voice_mail_plan'])
teste['churn'] = label_encoder.fit_transform(teste['churn'])
```

```python
teste.dtypes
```

<b>Etapa 4: Construindo Modelo de Machine Learning</b>

```python
# Import dos módulos
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

# Carregando os dados
array = dados.values
array2 = teste.values

# Separando o array em componentes de input e output
X_treino = array[:,0:19]
Y_treino = array[:,19]
X_teste = array2[:,0:19]
Y_teste = array2[:,19]

# Definindo os valores para o número de folds
#num_folds = 10
#seed = 7

# Separando os dados em folds
#kfold = KFold(num_folds, True, random_state = seed)

# Criando o modelo
modelo = LogisticRegression(solver='liblinear')

# Treinamento do modelo
modelo.fit(X_treino, Y_treino)

# Score do modelo nos dados de teste
result = modelo.score(X_teste, Y_teste)
print("Acurácia nos Dados de Teste: %.3f%%\n" % (result * 100.0))

# Fazendo as previsões e construindo a Confusion Matrix
previsoes = modelo.predict(X_teste)
matrix = confusion_matrix(Y_teste, previsoes)

# Imprimindo a Confusion Matrix
print("Confusion Matrix:")
print(matrix)

# Cross Validation
#resultado = cross_val_score(modelo, X, Y, cv = kfold, scoring = 'accuracy')

# Print dos resultados
#print("Acurácia: %.3f" % (resultado.mean() * 100))
```
